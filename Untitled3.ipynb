```python
# Tokenizations
import seaborn as sns
import matplotlib.pyplot as plt
import nltk
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.preprocessing import LabelBinarizer
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from wordcloud import WordCloud,STOPWORDS
from nltk.stem import WordNetLemmatizer
from nltk.tokenize import word_tokenize,sent_tokenize
import spacy
import re,string,unicodedata
from nltk.tokenize.toktok import ToktokTokenizer
from nltk.stem import LancasterStemmer,WordNetLemmatizer
from sklearn.linear_model import LogisticRegression,SGDClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from textblob import TextBlob
from textblob import Word
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from bs4 import BeautifulSoup

```

```python
import nltk
nltk.download('stopwords')
```

```python
# Tokenization of text
tokenizers = ToktokTokenizer()
#Setting English stopwords
stopwords = nltk.corpus.stopwords.words('english')
```

```python
#Removing the noisy text
def noiseremoval_text(text):
  soup = BeautifulSoup(text,"html.parser")
  text = soup.get_text()
  text = re.sub('\[[^]]*\]' , '' ,text)
  return text
```

```python
#Apply function on review column
data['review'] = data['review'].apply(noiseremoval_text)
```

```python
data.head()
```

```python
# Stemming the text
def stemmer(text):
  ps=nltk.porter.PorterStemmer()
  text = ' '.join([ps.stem(word) for word in text.split()])
  return text
```

```python
#Apply function on review column
data['review'] = data['review'].apply(stemmer)
```

```python
data.head()
```

```python
# removing the stopwords
def removing_stopwords(text , is_lower_case=False):
  #Tokenization of text
  tokenizers =ToktokTokenizer()
  #Setting Engish stopwords
  tokens = tokenizers.tokenize(text)
  tokens =[i.strip() for i in tokens]
  if is_lower_case:
    filtokens = [i for i in tokens if tokens not in  stopwords]
  else:
      filtokens = [i for i in tokens if i.lower() not in stopwords]

      filtered_texts =' '.join(filtokens)
      return filtered_texts
```

```python
data['review'] = data['review'].apply(removing_stopwords)
```

```python
data.head()
```

```python
#split the dataset
#train dataset
train_reviews_data = data.review[:30000]
```

```python
# test dataset
test_reviews_data = data.review[30000:]
```

```python
# count vectorizer for bag of words
cv = CountVectorizer(min_df=0,max_df =1,binary =False, ngram_range=(1,3))
# transform train reviews
cv_train = cv.fit_transform(train_reviews_data)
# transform test reviews
cv_test = cv.transform(test_reviews_data)

print("Bow_cv_train: " , cv_train.shape)
print("Bow_cv_test: " , cv_test.shape)
```

```python
#TF_IDF vectorizer
tf= TfidfVectorizer(min_df=0,max_df =1,use_idf =True , ngram_range=(1,3))
#transform test reviews
tf_train = tf.fit_transform(train_reviews_data)
# transformed test reviews
tf_test = tf.transform(test_reviews_data)
print("Tfidf_cv_train: " , tf_train.shape)
print("Tfidf_cv_test: " , tf_test.shape)

```

```python
#lebeling the sentiment data
label =LabelBinarizer()
#transform sentiment data
sentiment_data =label.fit_transform(data['sentiment'])
print(sentiment_data.shape)
```

```python
train_data = data.sentiment[:30000]
```

```python
test_data = data.sentiment[30000:]
```

```python
#training the model
logistic= LogisticRegression(penalty='l2', max_iter = 500, C=1,random_state=42)
#fitting the model for bag of words
lr_bow = logistic.fit(cv_train, train_data)
print(lr_bow)
```

```python
#prediction the model for bag of words
bow_predict = logistic.predict(cv_test)
print(bow_predict)
```

```python
#Accuracy score for bag of words
bow = accuracy_score(test_data,bow_predict)
print(bow)
```

```python
# Fitting he model for tfidf features
lr_tfidf = logistic.fit(tf_train,train_data)
print(lr_tfidf)
```

```python
#predicting
lr_tfidf_pre = logistic.predict(tf_test)
print(lr_tfidf_pre)
```

```python
#Accuracy score for tfidf feature
lr_tfidf_score = accuracy_score(test_data, lr_tfidf_pre)
print(lr_tfidf_score)
```
